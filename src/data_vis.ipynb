pipe

```python
PATH = "/home/subhankar/data/pipe/pipe"
INPUT_X = PATH +'/Pipe_X.npy'
INPUT_Y = PATH + '/Pipe_Y.npy'
OUTPUT_Sigma = PATH + '/Pipe_Q.npy'
pos_x = torch.tensor(np.load(INPUT_X), dtype=torch.float)
pos_y = torch.tensor(np.load(INPUT_Y), dtype=torch.float)
y = torch.tensor(np.load(OUTPUT_Sigma), dtype=torch.float)[:, 0].flatten(1,2)
pos = torch.stack([pos_x, pos_y], dim=-1).flatten(1,2)
```

darcy

```python
PATH = "/home/subhankar/data/darcy/Darcy_421/"
PATH_TRAIN = PATH +'piececonst_r421_N1024_smooth1.mat'
reader = MatReader(PATH_TRAIN)
r =5
train_a = reader.read_field('coeff')[:,::r,::r]
train_u = reader.read_field('sol')[:,::r,::r]
s = train_u.shape[1]

```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
x = np.linspace(0, 1, s)
y = np.linspace(0, 1, s)

# Create the 2D grid
X, Y = np.meshgrid(x, y)
ax[0].pcolormesh(X, Y, train_a[index].numpy(), shading='gouraud')
ax[1].pcolormesh(X, Y, train_u[index].numpy(), shading='gouraud')
```
